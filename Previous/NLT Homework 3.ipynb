1. Choose a English corpus from NLTK. Identify all the unique words in your corpus (types). This is your program's vocabulary. Compute the length of your corpus (N): total number of words in the corpus (tokens).

We first import the corpus of our choosing. A relatively short corpus is previous president Barak Obama's 2013 inaugural speech.

```python
import nltk
from nltk.corpus import inaugural

nltk.download('inaugural')
obama_speech = inaugural.words(fileids='2013-Obama.txt')
```

The corpus reader provided with NLTK has a variety of methods to read data from a corpus. In our case above, we are obtaining a list of words, but below are some examples that can read the corpus as raw text, a list of sentences, or a list of paragraphs.

```python
%%script echo Skipping
inaugural.raw('1789-Washington.txt')
# 'Fellow-Citizens of the Senate ...'
inaugural.sents('1789-Washington.txt')
# [['Fellow', '-', 'Citizens'...], ['Among', 'the', 'vicissitudes'...]...]
inaugural.paras('1789-Washington.txt')
# [[['Fellow', '-', 'Citizens'...]],
# [['Among', 'the', 'vicissitudes'...],
# ['On', 'the', 'one', 'hand', ',', 'I'...]...]...]
```

To determine the number of unique words in the corpus (types), we can use the `set()` function to convert a Python iterable, like a list, into a set data type. A set is a collection of unique objects that is unordered, therefore a set of the inaugural speech will only contain unique types of the corpus.

We can also count the length of the corpus by calling Python's `len()` function on the provided list.

```python
types = set(obama_speech)

tokens = len(obama_speech)

print("President Barrak Obama's 2013 Inaugural Speech")
print("Types:", len(types))
print("Tokens:", tokens)
```

2. Find the number of occurrences (frequency) of each type.

There are two ways to approach this problem. We can use the `Counter` class, a subclass of the dictionary, which will find the number of occurrences of each type. But I'm sure Dr. Kulkarni would want us to try it without using `Counter` too. I'll provide both, but my notebook will keep the results of the `Counter` class.

```python
from collections import Counter

word_freq = Counter(obama_speech)

for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
%%script echo Skipping
word_freq = {}
for word in obama_speech:
    if word in types:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

# Print the frequency of each word
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

3. Find the relative frequency of each type, $ W $, where relative frequency of `W = frequency_of_W / N`. This relative frequency is the probability (likelihood) of each type in the corpus.

For every word in the corpus, divide the frequency of the word by the total number of tokens. For example for the word "Thank" that occurs 3 times, the relative frequency is calculated by $ w = 3 / 2369 = 0.0012663571 $. Keep in mind that tokens like `.`, `:`, and the coma `,` are counted as unique types.

```python
relative_freq = {}
for word, freq in word_freq.items():
    relative_freq[word] = freq / tokens

# Print the relative frequency of each word
for word, rel_freq in relative_freq.items():
    print(f"{word}: {rel_freq}")
```

4. For every input string XYZ:

    4.a. If the input string XYZ exists in your vocabulary, return "XYZ is a complete and correct word as per corpus ___, and its probability is __"

    4.b. If the input string doesn't exist in your vocabulary, perform the below steps
    
    4.b.i. Calculate the similarity between the input string and each type using Levenshtein distance.

    4.b.ii. Output 5 types that are most similar to the input string as per the Levenshtein distance. Also, output the probability for each of the 5 types.

Firstly, lets write a function that takes the user input XYZ and calculates its similarity to every type according to the Levenshtein distance. First we'll import the Levenshtein Python C extension module.

```python
import Levenshtein
```

The Levenshtein module makes this task really. I call the `distance()` function on every type in the corpus and insert the result as a tuple of the word and edit distance into a list. I then sort the resulting list and output the top 5 (this value can be changed at the time of calling this function) similiars types to the input string.

```python
def get_most_similar_words(input_word, types, n=5):
    similar_words = []
    for word in types:
        distance = Levenshtein.distance(input_word, word)
        similar_words.append((word, distance))
    similar_words.sort(key=lambda x: x[1])
    return similar_words[:n]
```

I then take the user input, sanitize it by removing leading or trailing whitespace, and then check whether the user input is already in the collection of types. If it exists, I print that it exists in the corpus and its relative probability. If it does not exist, I print the five most similar types and the edit distance.

```python
user_input = input("Enter a word: ").strip()

if user_input in types:
    probability = relative_freq[user_input]
    print(f"{user_input} is a complete and correct word as per corpus \"2013-obama.txt\", and its probability is {probability}")
else:
    similar_words = get_most_similar_words(user_input, types)
    print(f"User input: {user_input}")
    for word, distance in similar_words:
        probability = relative_freq[word]
        print(f"{word} (Similarity: {distance}, Probability: {probability})")
```
